```python
import numpy as np
import matplotlib.pyplot as plt

# Define x1 values for plotting the lines
x1 = np.linspace(0, 8, 400)

# Constraint lines
x2_1 = (7 - x1) / 3         # x1 + 3x2 <= 7
x2_2 = 4 - x1               # 2x1 + 2x2 <= 8
x2_3 = 3 - x1               # x1 + x2 <= 3
x2_4 = 2 * np.ones_like(x1) # x2 <= 2
x2_min = np.zeros_like(x1)  # x2 >= 0

# Plot the constraint lines
plt.figure(figsize=(8,8))
plt.plot(x1, x2_1, label=r'$x_1 + 3x_2 \leq 7$')
plt.plot(x1, x2_2, label=r'$2x_1 + 2x_2 \leq 8$')
plt.plot(x1, x2_3, label=r'$x_1 + x_2 \leq 3$')
plt.plot(x1, x2_4, label=r'$x_2 \leq 2$')

# Create a grid of points to evaluate the feasible region
X1, X2 = np.meshgrid(np.linspace(0,8,400), np.linspace(0,5,400))

# Constraint conditions
cond1 = X1 + 3*X2 <= 7 + 1e-5
cond2 = 2*X1 + 2*X2 <= 8 + 1e-5
cond3 = X1 + X2 <= 3 + 1e-5
cond4 = X2 <= 2 + 1e-5
cond5 = X1 >= 0
cond6 = X2 >= 0

# Feasible region
region = cond1 & cond2 & cond3 & cond4 & cond5 & cond6

# Plot the feasible region
plt.imshow(region.astype(int), extent=(0,8,0,5), origin='lower', cmap='Greys', alpha=0.3)

# Vertices of the feasible region (precomputed)
vertices = np.array([
    [0,0],
    [0,2],
    [1,2],
    [2,0],
    [3,0]
])

plt.scatter(vertices[:,0], vertices[:,1], color='red')
for i, (x, y) in enumerate(vertices):
    plt.text(x + 0.1, y, f'({x},{y})', fontsize=9)

# Highlight optimal point
plt.scatter(3, 0, color='green', s=100, label='Optimal (3,0)')

# Plot formatting
plt.xlim(0,8)
plt.ylim(0,5)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title('Feasible Region and Constraints')
plt.legend()
plt.grid(True)
plt.show()
```
